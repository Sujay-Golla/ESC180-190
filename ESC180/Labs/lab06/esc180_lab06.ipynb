```python
# Problem 2
def get_nums(L):
  nums = []
  for i in L:
    nums.append(i[1])
  return nums

print(get_nums(L))
```

```python
# Problem 3
def lookup(L, num):
  for i in L:
    if i[1] == num:
      return i[0]

print(lookup(L, 99))
```
