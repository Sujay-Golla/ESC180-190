```python
import numpy as np
import matplotlib.pyplot as plt
def generate_data_noisy(a, num_points):
  x = np.linspace(0, 2 * np.pi, num_points)
  y = np.sin(a * x) + 0.1 * np.random.randn(num_points)
  return y
def generate_data(a, num_points):
  x = np.linspace(0, 2 * np.pi, num_points)
  y = np.sin(a * x)
  return y
def sum_squared_differences(y1, y2):
  s = 0
  for i in range(len(y1)):
    s = s + (y1[i] - y2[i]) ** 2
  return s
if __name__ == "__main__":
  secret_a = 2.5
  x = np.linspace(0, 2 * np.pi, 100)
  y_noise = generate_data_noisy(secret_a, 100)
  min_diff = 10**100
  min_a = 10**100

  i = 0
  while i <= 5:
      y = generate_data(i, 100)
      temp = sum_squared_differences(y, y_noise)
      if temp < min_diff:
        min_a = i
        min_diff = temp
      i += 0.001

  y = generate_data(min_a, 100)

  print("diff:", sum_squared_differences(y, y_noise), "\na:", min_a)
  plt.plot(x, y)
  plt.plot(x, y_noise)
  plt.show()
# Your code here: try different aâ€™s and find the one that makes
# sum_squared_differences(y, generate_data(a, len(y))) small.
# Your code here: plot y together with the best fit sin(ax).
```

```python
import random
def generate_hats(num_students):
  hats = []
  for i in range(num_students):
    if random.random() < 0.5:
      hats.append("W")
    else:
      hats.append("B")
  return hats
def count_white_hats(hats, i):
  counter = 0
  for j in range(0, len(hats)):
    if j != i and hats[j] == "W":
      counter += 1
  return counter
def make_guesses(hats):
  # Return a list of guesses, one for each hat in hats.
  # The guesses should be â€™Wâ€™ or â€™Bâ€™.
  # Your code here
  # For the student hats[i], make a guess based on the number
  # of white hats that they see
  # You can use count_white_hats() to help you figure out the
  # number of white hats that a student sees
  guesses = []
  for i in range(0,len(hats)):
    count = count_white_hats(hats, i)
    if count % 2 == 0: # if everyone guesses there's even number of white hats
      guesses.append("B")
    else:
      guesses.append("W")
  return guesses
def are_all_guesses_correct(hats, guesses):
# Return True if all guesses are correct.
# Your code here
  for i in range(0, len(hats)):
    if hats[i] != guesses[i]:
      return False
  return True
if __name__ == "__main__":
  num_students = 12
  hats = generate_hats(num_students)
  print("Hats: ", hats)
  guesses = make_guesses(hats)
  print("Guesses:", guesses)
  if are_all_guesses_correct(hats, guesses):
    print("All guesses correct!")
  else:
    print("Not all guesses correct.")
```

```python
def gcd(n, m):
  gcd = 0
  for i in range(1, min(n, m)+1):
    if n%i == 0 and m%i == 0:
      gcd = i
  return gcd

if __name__ == "__main__":
  print(gcd(15, 10))
  print(gcd(13, 17))
  print(gcd(13, 26))
```
